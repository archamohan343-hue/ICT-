```python
def calculate_total_bill(item_price):
  platformfee=5
  deliverycharge=80
  subtotal=sum(item_price)
  subtotal+=platformfee
  subtotal+=deliverycharge
  return subtotal

item_price=[100,200,300,800]
subtotal= calculate_total_bill(item_price)
if subtotal>1000:
  reward=200
elif subtotal>500:
  reward=50
else:
  reward=0


print(f"Your subtotal is,{subtotal} and reward is {reward} ")
```
